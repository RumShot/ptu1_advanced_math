# Uzduotys 2

```python
pirmas = pd.read_csv("top1-25-1.csv")
antras = pd.read_csv("top1-25-2.csv")
trecias = pd.read_csv("top26-50-1.csv")
ketvirtas = pd.read_csv("top26-50-2.csv")
# ketvirtas
```

```python
cols_to_use = antras.columns.difference(pirmas.columns)
virsus = pd.merge(pirmas, antras[cols_to_use], left_index=True, right_index=True, how='outer')
virsus
```

```python
cols_to_use2 = ketvirtas.columns.difference(trecias.columns)
apacia = pd.merge(trecias, ketvirtas[cols_to_use2], left_index=True, right_index=True, how='outer')
apacia
```

```python
visa_lentele = pd.concat([virsus, apacia])
visa_lentele
```

```python
rs_index = visa_lentele.reset_index()
rs_index.drop(columns='index', inplace = True)
rs_index
```

```python
grouped = rs_index.groupby('Genre')
ze_type = type(grouped)
# ze_type
grouped
```

```python
repeated_genre = pd.pivot_table(rs_index, columns= ['Genre'], aggfunc='size')
repeated_genre[repeated_genre > 3]
```
