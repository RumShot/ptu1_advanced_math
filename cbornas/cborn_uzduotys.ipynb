```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

# %matplotlib inline

sns.set_style('darkgrid')

mpg = sns.load_dataset('mpg')
mpg.head(5)
```

# 1, 2 uzd

```python
mpg.shape
```

```python
# sns.set_style('darkgrid')
# sns.set_palette("pastel")
accel = sns.histplot(mpg['acceleration'], edgecolor = "white")
accel.set(ylabel=None)
```

```python
# sns.set_style('darkgrid')
# sns.set_palette("pastel")
displ = sns.histplot(mpg['displacement'], edgecolor = "white", palette="skyblue")
displ.set(ylabel=None)
```

# 3 uzd

```python
sns.set_style('darkgrid')
sns.histplot(x ='cylinders', data=mpg, hue='cylinders', edgecolor = "white",
 palette='summer', alpha=0.9, legend=None, bins=8)
```

```python
sns.countplot('cylinders', data=mpg)
```

```python
sns.countplot('origin', data=mpg)
```
