```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

# %matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.histplot(tips['total_bill'], bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sns.scatterplot('tip', 'total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', style='time')
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='size', size='size', sizes=(20, 200))
```

```python
sns.pairplot(tips, hue='sex', diag_kind='hist')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```
